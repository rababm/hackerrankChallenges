```python
import os
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegressionCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

firstLine = input().split(" ")
numberOfTrainingData = int(firstLine[0])
numberOfFeatures = int(firstLine[1])

def fetchTrainingData() :
    labels = []
    y_train = []
    X_train = np.empty(shape=[numberOfTrainingData, numberOfFeatures+1])
    for i in range(0, numberOfTrainingData):
        line = input().split(" ")
        identifier = line[0]
        y_train.append(line[1])
        line = line[2:] 
        for featureLine in line:
            feature = featureLine.split(":")
            featureNumber = int(feature[0])
            featureValue = float(feature[1])
            X_train[i][featureNumber-1] = featureValue
    return numberOfFeatures, X_train, y_train

numberOfFeatures, X_train, y_train = fetchTrainingData()

def fetchTestData(numberOfFeatures = numberOfFeatures) :
    numberOfTestData = int(input())
    X_test = np.empty(shape=[numberOfTestData, numberOfFeatures])
    identifier = []
    for i in range(0, numberOfTestData):
        line = input().split(" ")
        identifier.append(line[0])
        line = line[1:] 
        for featureLine in line:
            feature = featureLine.split(":")
            featureNumber = int(feature[0])
            featureValue = float(feature[1])
            X_test[i][featureNumber-1] = featureValue
    return identifier, X_test

numberOfLabels = len(y_train)
labelEncoder = preprocessing.LabelEncoder()
labels = labelEncoder.fit_transform(y_train)

for i in range(0,numberOfLabels):
    X_train[i, 23]=labels[i]
    
trainingData = pd.DataFrame(X_train)

def transformData(data):
    data.drop([18,21,22], axis=1, inplace=True)
    data[18]=data[13]+data[9]
    data.drop([13,9], axis=1, inplace=True)
    return data
    
def transformTrainingData(data = trainingData):
    data = data[trainingData[3] >2]
    data = data[trainingData[4] >2]
    
    return data
    
trainingData = transformTrainingData()

labels = np.array(trainingData[23])
trainingData.drop(23, axis=1, inplace=True)

trainingData = transformData(trainingData)

clf = RandomForestClassifier().fit(trainingData, labels)

identifier, X_test = fetchTestData()

testData = pd.DataFrame(np.array(X_test))
testData = transformData(testData)
    
testData.describe()
predicted = clf.predict(testData)
#print(predicted)
p = labelEncoder.inverse_transform(predicted.astype(int))
i=0
for label in p:
    print(identifier[i], label)
    i = i+1
```
